```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```
df = pd.read_csv('/content/titanic.csv')
```

```
sns.set_style("whitegrid")
```

```
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title("Histogram of Age")
plt.show()
```

```
plt.figure(figsize=(8, 5))
sns.barplot(x=df['Pclass'].value_counts().index, y=df['Pclass'].value_counts().values)
plt.title("Bar Chart of Passenger Class")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

```
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Pclass'], y=df['Age'])
plt.title("Box Plot of Age by Passenger Class")
plt.show()
```

```
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']].dropna(), hue="Survived")
plt.show()
```

```
plt.figure(figsize=(8, 5))
numerical_df = df.select_dtypes(include=['number'])
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```
plt.figure(figsize=(8, 5))
sns.violinplot(x=df['Pclass'], y=df['Age'], hue=df['Survived'], split=True)
plt.title("Violin Plot of Age by Passenger Class and Survival")
plt.show()

```

```
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Survived'])
plt.title("Count Plot of Survival")
plt.show()

```

```
plt.figure(figsize=(8, 5))
plt.plot(df.index, df['Fare'])
plt.title("Line Chart of Fare")
plt.xlabel("Passenger Index")
plt.ylabel("Fare")
plt.show()
```

```
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Age'], y=df['Fare'], hue=df['Survived'])
plt.title("Scatter Plot of Age vs Fare")
plt.show()
```

```
plt.figure(figsize=(8, 5))
sns.kdeplot(df['Age'].dropna(), shade=True)
plt.title("KDE Plot of Age")
plt.show()
```
